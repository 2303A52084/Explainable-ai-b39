```python
# ======================================================
# FULL DATA SCIENCE PIPELINE (PREPROCESSING → ML → DL)
# ======================================================
# Author: GPT-5 | All stages implemented

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import shap
import warnings
warnings.filterwarnings('ignore')

# ML + Preprocessing
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix, roc_curve, auc
from sklearn.utils import resample
from imblearn.over_sampling import SMOTE

# Machine Learning Models
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC

# Ensemble Methods
from sklearn.ensemble import (
    BaggingClassifier, RandomForestClassifier, ExtraTreesClassifier,
    AdaBoostClassifier, GradientBoostingClassifier, VotingClassifier, StackingClassifier
)
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from catboost import CatBoostClassifier

# Deep Learning (Feedforward + CNN + LSTM)
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Conv1D, MaxPooling1D, Flatten, LSTM
from tensorflow.keras.optimizers import Adam

# ======================================================
# LOAD DATA
# ======================================================
df = pd.read_csv("test (1).csv")
print("Initial Shape:", df.shape)

# ======================================================
# NULL VALUE HANDLING
# ======================================================
print("\nMissing values before:")
print(df.isnull().sum())

for col in df.columns:
    if df[col].dtype in ['int64', 'float64']:
        df[col].fillna(df[col].median(), inplace=True)
    else:
        df[col].fillna(df[col].mode()[0], inplace=True)

print("\nMissing values after:")
print(df.isnull().sum())

# ======================================================
# DETECT TARGET COLUMN
# ======================================================
target = df.columns[-1]
print(f"\nDetected Target Column: {target}")

# Separate features and target, dropping non-numeric columns like 'Patient Id'
X = df.drop(columns=[target, 'Patient Id', 'Patient First Name', 'Family Name', 'Father\'s name', 'Institute Name', 'Location of Institute', 'Status', 'Gender', 'Place of birth', 'Blood test result'])
y = df[target]

# Identify categorical columns to handle separately if needed later
categorical_cols = X.select_dtypes(include=['object', 'bool']).columns
print(f"\nIdentified Categorical Columns: {list(categorical_cols)}")

# Convert boolean columns to numeric (0 and 1)
for col in categorical_cols:
    if X[col].dtype == 'bool':
        X[col] = X[col].astype(int)


# ======================================================
# NORMALIZATION (Min-Max + Z-Score)
# ======================================================
# Ensure only numeric columns are selected for scaling
X_numeric = X.select_dtypes(include=np.number)

minmax_scaler = MinMaxScaler()
standard_scaler = StandardScaler()

X_minmax = pd.DataFrame(minmax_scaler.fit_transform(X_numeric), columns=X_numeric.columns)
X_scaled = pd.DataFrame(standard_scaler.fit_transform(X_minmax), columns=X_numeric.columns)

# ======================================================
# HANDLE CLASS IMBALANCE USING SMOTE
# ======================================================
if y.nunique() <= 20: # Assuming target is categorical or binary
    if y.dtype == 'object' or y.dtype == 'bool': # Check if target is not numeric for SMOTE
        # Convert target to numeric if it's not already
        y = y.astype('category').cat.codes
    sm = SMOTE(random_state=42)
    X_scaled, y = sm.fit_resample(X_scaled, y)
    print("\nApplied SMOTE for balancing class distribution.")
    print("New Class Distribution:\n", pd.Series(y).value_counts())
else:
    print("\nSMOTE not applied as target is not categorical or has too many unique values.")


# ======================================================
# CORRELATION ANALYSIS
# ======================================================
plt.figure(figsize=(10, 8))
sns.heatmap(X_scaled.corr(), cmap="coolwarm", center=0)
plt.title("Feature Correlation Heatmap")
plt.show()

corr_matrix = X_scaled.corr().abs()
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
to_drop = [c for c in upper.columns if any(upper[c] > 0.85)]
X_reduced = X_scaled.drop(columns=to_drop)
print("\nDropped highly correlated features:", to_drop)

# ======================================================
# TRAIN-TEST SPLIT
# ======================================================
X_train, X_test, y_train, y_test = train_test_split(X_reduced, y, test_size=0.2, random_state=42)

# ======================================================
# SHAP FEATURE IMPORTANCE
# ======================================================
# Ensure X_train has enough samples and features for SHAP
if X_train.shape[0] > 0 and X_train.shape[1] > 0:
    rf_model = RandomForestClassifier(random_state=42)
    rf_model.fit(X_train, y_train)
    explainer = shap.TreeExplainer(rf_model)
    # SHAP values for tree models can be calculated directly on the training data
    shap_values = explainer.shap_values(X_train)
    # If target is binary, shap_values will be a list with two arrays, take the one for the positive class
    if isinstance(shap_values, list):
        shap.summary_plot(shap_values[1], X_train, plot_type="bar")
    else:
        shap.summary_plot(shap_values, X_train, plot_type="bar")
else:
    print("\nSkipping SHAP Feature Importance: X_train is empty or has no features.")


# ======================================================
# VISUALIZATION TECHNIQUES
# ======================================================
# Histogram
# Select only numeric columns for histogram
X_numeric.hist(figsize=(12, 10))
plt.suptitle("Feature Distributions")
plt.show()

# Box Plot
plt.figure(figsize=(10, 6))
# Select only numeric columns for boxplot
sns.boxplot(data=X_numeric)
plt.title("Boxplot for Outlier Detection")
plt.xticks(rotation=90) # Rotate x-axis labels for better readability
plt.tight_layout() # Adjust layout to prevent labels overlapping
plt.show()

# Pair Plot
# Select a subset of numeric columns for pairplot to avoid excessive plot generation
numeric_cols_subset = X_numeric.columns[:5] # Adjust the number of columns as needed
if len(numeric_cols_subset) > 0:
    sns.pairplot(pd.concat([X_numeric[numeric_cols_subset], y], axis=1), diag_kind="kde")
    plt.show()
else:
    print("\nSkipping Pair Plot: No numeric columns available for plotting.")


# Violin Plot (for numeric variables)
# Use the same subset of numeric columns as for pairplot
if len(numeric_cols_subset) > 0:
    for col in numeric_cols_subset:
        plt.figure(figsize=(6, 4))
        sns.violinplot(x=y, y=X_numeric[col])
        plt.title(f"Violin Plot - {col}")
        plt.show()
else:
     print("\nSkipping Violin Plot: No numeric columns available for plotting.")


# ======================================================
# MACHINE LEARNING MODELS
# ======================================================
models = {
    "Logistic Regression": LogisticRegression(max_iter=1000),
    "Decision Tree": DecisionTreeClassifier(),
    "KNN": KNeighborsClassifier(),
    "Naive Bayes": GaussianNB(),
    "SVM": SVC(probability=True)
}

results = {}

for name, clf in models.items():
    # Check if the model supports predict_proba for AUC calculation
    needs_proba = hasattr(clf, "predict_proba") and y.nunique() == 2 # AUC is typically for binary classification
    try:
        clf.fit(X_train, y_train)
        y_pred = clf.predict(X_test)
        y_prob = clf.predict_proba(X_test)[:, 1] if needs_proba else None

        acc = accuracy_score(y_test, y_pred)
        # Use zero_division parameter for precision, recall, and f1 score
        prec = precision_score(y_test, y_pred, average='weighted', zero_division=0)
        rec = recall_score(y_test, y_pred, average='weighted', zero_division=0)
        f1 = f1_score(y_test, y_pred, average='weighted', zero_division=0)
        auc_val = roc_auc_score(y_test, y_prob) if y_prob is not None else None
        results[name] = [acc, prec, rec, f1, auc_val]
    except Exception as e:
        print(f"Error training or evaluating {name}: {e}")
        results[name] = [None, None, None, None, None]


# ======================================================
# ENSEMBLE MODELS
# ======================================================
ensemble_models = {
    "Bagging": BaggingClassifier(random_state=42),
    "Random Forest": RandomForestClassifier(random_state=42),
    "Extra Trees": ExtraTreesClassifier(random_state=42),
    "AdaBoost": AdaBoostClassifier(random_state=42),
    "Gradient Boosting": GradientBoostingClassifier(random_state=42),
    "XGBoost": XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42),
    "LightGBM": LGBMClassifier(random_state=42),
    "CatBoost": CatBoostClassifier(verbose=0, random_state=42)
}

for name, clf in ensemble_models.items():
    needs_proba = hasattr(clf, "predict_proba") and y.nunique() == 2
    try:
        clf.fit(X_train, y_train)
        y_pred = clf.predict(X_test)
        y_prob = clf.predict_proba(X_test)[:, 1] if needs_proba else None

        acc = accuracy_score(y_test, y_pred)
        prec = precision_score(y_test, y_pred, average='weighted', zero_division=0)
        rec = recall_score(y_test, y_pred, average='weighted', zero_division=0)
        f1 = f1_score(y_test, y_pred, average='weighted', zero_division=0)
        auc_val = roc_auc_score(y_test, y_prob) if y_prob is not None else None
        results[name] = [acc, prec, rec, f1, auc_val]
    except Exception as e:
        print(f"Error training or evaluating {name}: {e}")
        results[name] = [None, None, None, None, None]


# ======================================================
# STACKING & VOTING CLASSIFIERS
# ======================================================
# Ensure base estimators are compatible with the data and task
estimators = [
    ('lr', LogisticRegression(max_iter=1000, random_state=42)),
    ('rf', RandomForestClassifier(random_state=42)),
    ('svc', SVC(probability=True, random_state=42))
]

# Check if all estimators support predict_proba for VotingClassifier with voting='soft'
if all(hasattr(estimator[1], 'predict_proba') for estimator in estimators):
    stack_model = StackingClassifier(estimators=estimators, final_estimator=LogisticRegression(random_state=42))
    voting_model = VotingClassifier(estimators=estimators, voting='soft')

    for name, clf in {"Stacking": stack_model, "Voting": voting_model}.items():
        try:
            clf.fit(X_train, y_train)
            y_pred = clf.predict(X_test)
            acc = accuracy_score(y_test, y_pred)
            # Stacking and Voting performance metrics can be calculated similarly to other models if needed
            # For simplicity, only accuracy is shown here as in the original code
            results[name] = [acc, None, None, None, None]
        except Exception as e:
            print(f"Error training or evaluating {name}: {e}")
            results[name] = [None, None, None, None, None]
else:
    print("\nSkipping Stacking and Voting Classifiers: Not all base estimators support predict_proba.")


# ======================================================
# DEEP LEARNING MODELS
# ======================================================
# Ensure data is in the correct format for deep learning models
X_train_dl = X_train.values
X_test_dl = X_test.values

# Feedforward Neural Network
if X_train_dl.shape[0] > 0 and X_train_dl.shape[1] > 0:
    model_ffn = Sequential([
        Dense(128, input_dim=X_train_dl.shape[1], activation='relu'),
        Dropout(0.3),
        Dense(64, activation='relu'),
        Dense(1, activation='sigmoid') # Assuming binary classification
    ])
    model_ffn.compile(optimizer=Adam(0.001), loss='binary_crossentropy', metrics=['accuracy'])
    try:
        history = model_ffn.fit(X_train_dl, y_train, epochs=15, batch_size=32, validation_split=0.2, verbose=0)

        plt.plot(history.history['accuracy'], label='Train Acc')
        plt.plot(history.history['val_accuracy'], label='Val Acc')
        plt.legend()
        plt.title("Feedforward NN Accuracy")
        plt.show()
    except Exception as e:
        print(f"Error training Feedforward Neural Network: {e}")
else:
    print("\nSkipping Feedforward Neural Network: X_train is empty or has no features.")


# CNN (1D)
# Reshape data for CNN (samples, timesteps, features)
X_train_cnn = np.expand_dims(X_train_dl, axis=2)
X_test_cnn = np.expand_dims(X_test_dl, axis=2)

if X_train_cnn.shape[0] > 0 and X_train_cnn.shape[1] > 0:
    cnn = Sequential([
        Conv1D(64, 3, activation='relu', input_shape=(X_train_cnn.shape[1], 1)),
        MaxPooling1D(2),
        Flatten(),
        Dense(64, activation='relu'),
        Dense(1, activation='sigmoid') # Assuming binary classification
    ])
    cnn.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    try:
        cnn.fit(X_train_cnn, y_train, epochs=10, batch_size=32, verbose=0, validation_split=0.2)
    except Exception as e:
        print(f"Error training CNN (1D): {e}")
else:
    print("\nSkipping CNN (1D): X_train is empty or has no features.")


# LSTM
# Reshape data for LSTM (samples, timesteps, features) - same as CNN 1D in this case
if X_train_cnn.shape[0] > 0 and X_train_cnn.shape[1] > 0:
    lstm = Sequential([
        LSTM(64, input_shape=(X_train_cnn.shape[1], 1)),
        Dense(1, activation='sigmoid') # Assuming binary classification
    ])
    lstm.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    try:
        lstm.fit(X_train_cnn, y_train, epochs=10, batch_size=32, verbose=0, validation_split=0.2)
    except Exception as e:
        print(f"Error training LSTM: {e}")
else:
    print("\nSkipping LSTM: X_train is empty or has no features.")


# ======================================================
# CONFUSION MATRIX & ROC
# ======================================================
# Ensure X_test has enough samples and features for prediction
if X_test.shape[0] > 0 and X_test.shape[1] > 0:
    # Use the best performing model from the summary or a chosen model
    # Here, using RandomForestClassifier as in the original code
    best_model = RandomForestClassifier(random_state=42).fit(X_train, y_train)
    y_pred = best_model.predict(X_test)
    cm = confusion_matrix(y_test, y_pred)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title("Confusion Matrix - Random Forest")
    plt.show()

    if hasattr(best_model, "predict_proba") and y.nunique() == 2:
        try:
            y_prob = best_model.predict_proba(X_test)[:, 1]
            fpr, tpr, _ = roc_curve(y_test, y_prob)
            plt.plot(fpr, tpr, label=f"AUC={auc(fpr, tpr):.2f}")
            plt.plot([0, 1], [0, 1], 'k--')
            plt.legend()
            plt.title("ROC Curve - Random Forest")
            plt.show()
        except Exception as e:
            print(f"Error generating ROC Curve: {e}")
    else:
        print("\nSkipping ROC Curve: Model does not support predict_proba or target is not binary.")
else:
    print("\nSkipping Confusion Matrix and ROC Curve: X_test is empty or has no features.")


# ======================================================
# SUMMARY RESULTS
# ======================================================
results_df = pd.DataFrame(results, index=["Accuracy", "Precision", "Recall", "F1", "AUC"]).T
print("\n=== MODEL PERFORMANCE SUMMARY ===")
print(results_df.sort_values("Accuracy", ascending=False))
```

```python
# ======================================================
# 🔍 CORRELATION ANALYSIS + ⚙️ SHAP FEATURE IMPORTANCE
# ======================================================

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import shap
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

pd.set_option('future.no_silent_downcasting', True)  # avoid warnings

print("\n🚀 Starting Correlation + SHAP Analysis...")

# --------------------------------------------
# 1️⃣ Ensure X and y exist (feature/target split)
# --------------------------------------------
if 'X' not in locals() or 'y' not in locals():
    print("⚠️ X or y not found. Creating from df...")

    target = df.columns[-1]  # last column as target
    y = df[target]

    # Drop unwanted or ID columns safely
    drop_cols = [
        target, 'Patient Id', 'Patient First Name', 'Family Name', "Father's name",
        'Institute Name', 'Location of Institute', 'Status', 'Gender',
        'Place of birth', 'Blood test result'
    ]
    drop_cols = [c for c in drop_cols if c in df.columns]

    X = df.drop(columns=drop_cols)

    print(f"✅ Created X (features) and y (target = '{target}')")

# --------------------------------------------
# 2️⃣ Handle missing values (numeric + categorical)
# --------------------------------------------
print("Handling missing values in X...")
for col in X.columns:
    if X[col].dtype in ['int64', 'float64']:
        X[col] = X[col].fillna(X[col].median())
    else:
        X[col] = X[col].fillna(X[col].mode()[0])

print("Handling missing values in y...")
if y.dtype in ['int64', 'float64']:
    y = y.fillna(y.median())
else:
    y = y.fillna(y.mode()[0])

print("✅ Missing values handled.")

# --------------------------------------------
# Convert target to numeric for classification
if y.dtype == 'object' or y.dtype == 'bool':
    print("Converting target variable to numeric...")
    y = y.astype('category').cat.codes
    print("✅ Target variable converted.")

# 3️⃣ Create scaled data (X_scaled)
# --------------------------------------------
if 'X_scaled' not in locals():
    print("⚙️ Scaling numeric columns...")
    X_numeric = X.select_dtypes(include=np.number)

    minmax_scaler = MinMaxScaler()
    standard_scaler = StandardScaler()

    X_minmax = pd.DataFrame(minmax_scaler.fit_transform(X_numeric), columns=X_numeric.columns)
    X_scaled = pd.DataFrame(standard_scaler.fit_transform(X_minmax), columns=X_numeric.columns)

    print(f"✅ X_scaled created with shape {X_scaled.shape}")

# --------------------------------------------
# 4️⃣ Perform Correlation Analysis
# --------------------------------------------
print("\n🔍 Performing Correlation Analysis...")

corr_matrix = X_scaled.corr().abs()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, cmap="coolwarm", center=0)
plt.title("Feature Correlation Heatmap (to detect multicollinearity)")
plt.show()

upper_triangle = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
to_drop = [col for col in upper_triangle.columns if any(upper_triangle[col] > 0.85)]

X_reduced = X_scaled.drop(columns=to_drop, errors='ignore')
print(f"✅ Dropped highly correlated features (r > 0.85): {to_drop}")
print(f"Shape before: {X_scaled.shape} | Shape after: {X_reduced.shape}")

# --------------------------------------------
# 5️⃣ Train/Test Split
# --------------------------------------------
X_train, X_test, y_train, y_test = train_test_split(X_reduced, y, test_size=0.2, random_state=42)

# Add checks for NaNs before training
if X_train.isnull().sum().sum() > 0:
    print("⚠️ NaNs found in X_train BEFORE training. Investigating...")
    # You might want to inspect specific columns here: X_train.isnull().sum()
    # For now, I'll attempt to impute again just in case
    for col in X_train.columns:
         if X_train[col].dtype in ['int64', 'float64']:
             X_train[col] = X_train[col].fillna(X_train[col].median())
         else:
             X_train[col] = X_train[col].fillna(X_train[col].mode()[0])
    print("✅ Attempted to re-impute NaNs in X_train.")


if y_train.isnull().sum() > 0:
     print("⚠️ NaNs found in y_train BEFORE training. Investigating...")
     # For now, I'll attempt to impute again just in case
     if y_train.dtype in ['int64', 'float64']:
         y_train = y_train.fillna(y_train.median())
     else:
         y_train = y_train.fillna(y_train.mode()[0])
     print("✅ Attempted to re-impute NaNs in y_train.")


# --------------------------------------------
# 6️⃣ SHAP Feature Importance
# --------------------------------------------
print("\n⚙️ Computing SHAP Feature Importance...")

rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train, y_train)

explainer = shap.TreeExplainer(rf_model)
shap_values = explainer.shap_values(X_train)

print("Generating SHAP Summary Plot...")
plt.title("SHAP Feature Importance - Global Explanation")

if isinstance(shap_values, list):  # binary classification
    shap.summary_plot(shap_values[1], X_train, plot_type="bar", show=True)
else:
    shap.summary_plot(shap_values, X_train, plot_type="bar", show=True)

print("\n✅ SHAP Feature Importance analysis completed successfully.")
```
